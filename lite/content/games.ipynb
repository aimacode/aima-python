# Adversarial search in the browser

Minimax and alpha-beta on Tic-Tac-Toe from chapter 5, running fully in your
browser with `aima.games`.

```python
import piplite
await piplite.install("aima", deps=False)
```

```python
from aima.games import TicTacToe, minmax_decision, alpha_beta_search

game = TicTacToe()
state = game.initial
print("minimax optimal first move:   ", minmax_decision(state, game))
print("alpha-beta optimal first move:", alpha_beta_search(state, game))
```

```python
import random
from aima.games import alpha_beta_player, random_player

# An optimal alpha-beta player never loses: against a random opponent it wins
# or draws every game (utility >= 0 for the first player).
random.seed(0)
results = [game.play_game(alpha_beta_player, random_player) for _ in range(10)]
print("alpha-beta vs random, 10 games (first-player utility):", results)
print("never loses:", all(u >= 0 for u in results))
```
