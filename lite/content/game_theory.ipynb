# Game theory in the browser

Pure-strategy Nash equilibria, zero-sum games and the Shapley value (chapter 18), running fully in your browser with `aima.game_theory`.

`aima.game_theory` uses SciPy for the linear program behind zero-sum games, so we load it alongside `aima`.

```python
import piplite
await piplite.install("scipy")
await piplite.install("aima", deps=False)
```

## Pure-strategy Nash equilibria

Payoffs as utilities, indexed `[row player][column player]`.

```python
from aima.game_theory import pure_nash_equilibria, solve_zero_sum_game, shapley_value

# prisoner's dilemma (0 = testify, 1 = refuse)
ALI = [[-5, 0], [-10, -1]]
BO = [[-5, -10], [0, -1]]
print('prisoner\'s dilemma Nash:', pure_nash_equilibria(ALI, BO))
print('matching pennies Nash:  ', pure_nash_equilibria([[1, -1], [-1, 1]], [[-1, 1], [1, -1]]))
print('coordination game Nash: ', pure_nash_equilibria([[2, 0], [0, 1]], [[2, 0], [0, 1]]))
```

## Zero-sum games and cooperative value

Two-finger Morra is a zero-sum game with value -1/12; the Shapley value splits the gains of the three-player glove game fairly.

```python
value, row, col = solve_zero_sum_game([[2, -3], [-3, 4]])
print('two-finger Morra value:', round(value, 4), ' row strategy:', [round(x, 3) for x in row])

# glove game: players 1,2 own a left glove each, player 3 a right glove;
# a pair is worth 1
def gloves(coalition):
    left = {1, 2} & set(coalition)
    right = {3} & set(coalition)
    return min(len(left), len(right))
print('Shapley value of the glove game:', shapley_value([1, 2, 3], gloves))
```
