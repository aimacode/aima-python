# Reinforcement learning in the browser

Q-learning on the 4x3 grid world (chapter 22), running fully in your browser with `aima.reinforcement_learning`. The agent has no model of the environment and learns purely from experience.

```python
import piplite
await piplite.install("aima", deps=False)
```

```python
import random
from aima.mdp import sequential_decision_environment as grid
from aima.reinforcement_learning import QLearningAgent, run_single_trial

random.seed(0)
agent = QLearningAgent(grid, Ne=5, Rplus=2,
                       alpha=lambda n: 60 / (59 + n))
for _ in range(300):
    run_single_trial(agent, grid)

# derive a greedy policy from the learned action-value function Q
U = {}
for (state, action), q in agent.Q.items():
    if q > U.get(state, float('-inf')):
        U[state] = q
print('learned', len(agent.Q), 'state-action values over 300 trials')
print('estimated utility of the start state (0, 0):',
      round(max(agent.Q[((0, 0), a)] for a in grid.actions((0, 0))), 3))
```
