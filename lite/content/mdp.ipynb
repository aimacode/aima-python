## The 4x3 grid world

An agent moves on a grid with a +1 goal, a -1 trap and a wall; actions succeed 80% of the time. Value iteration computes the utility of every state, and `best_policy` reads off the optimal action.

```python
from aima.mdp import sequential_decision_environment as grid
from aima.mdp import value_iteration, best_policy, policy_iteration

U = value_iteration(grid)
pi = best_policy(grid, U)
print('optimal policy (value iteration):')
print('\n'.join(' '.join(c if c is not None else ' ' for c in row) for row in grid.to_arrows(pi)))
```

Policy iteration converges to the same optimal policy by a different route.

```python
pi2 = policy_iteration(grid)
print('optimal policy (policy iteration):')
print('\n'.join(' '.join(c if c is not None else ' ' for c in row) for row in grid.to_arrows(pi2)))
```
