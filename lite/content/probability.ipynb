# Probabilistic reasoning in the browser

Exact and approximate inference on a Bayesian network (chapters 12-13),
running fully in your browser with `aima.probability`.

```python
import piplite
await piplite.install("aima", deps=False)
```

## The burglary network

The classic alarm network (Figure 12.2): a burglary or an earthquake may set
off the alarm, which John and Mary may then report. Given that **both** John
and Mary call, how likely is a burglary?

```python
from aima.probability import burglary, enumeration_ask, elimination_ask

evidence = dict(JohnCalls=True, MaryCalls=True)
print("enumeration  P(Burglary | J, M):", enumeration_ask("Burglary", evidence, burglary).show_approx())
print("elimination  P(Burglary | J, M):", elimination_ask("Burglary", evidence, burglary).show_approx())
```

## Approximate inference

Both exact methods agree on P(Burglary | J, M) ≈ 0.284. Likelihood weighting
approximates the same posterior by sampling.

```python
import random
from aima.probability import likelihood_weighting

random.seed(42)
print("likelihood weighting (10000 samples):",
      likelihood_weighting("Burglary", evidence, burglary, 10000).show_approx())
```
