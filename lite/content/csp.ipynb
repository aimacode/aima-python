## Map colouring

Colour the map of Australia with three colours so that no two neighbouring
regions share one. First we enforce arc consistency (AC-3), then solve by
backtracking search.

```python
from aima.csp import australia_csp, AC3, backtracking_search

consistent, checks = AC3(australia_csp)
print("AC-3 arc-consistent:", consistent, "  (", checks, "constraint checks )")
print("backtracking solution:", backtracking_search(australia_csp))
```

## N-queens

Place 8 non-attacking queens by local search (min-conflicts). The solution
maps each column to the row of its queen.

```python
import random
from aima.csp import NQueensCSP, min_conflicts

random.seed(0)
solution = min_conflicts(NQueensCSP(8), max_steps=1000)
print("8-queens (column -> row):", solution)
```
