# Classical planning in the browser

STRIPS-style planning (chapter 11) solved with GraphPlan, running fully in
your browser with `aima.planning`.

```python
import piplite
await piplite.install("aima", deps=False)
```

## Spare tire

Replace a flat tire with the spare from the trunk. `GraphPlan` returns a
layered plan; `linearize` flattens it into an executable action sequence.

```python
from aima.planning import spare_tire, air_cargo, GraphPlan, linearize

plan = GraphPlan(spare_tire()).execute()
print("spare tire plan:", linearize(plan))
```

## Air cargo

Fly two cargos to their destinations (Figure 11.2) — a larger problem with
load/fly/unload actions across two airports.

```python
plan = GraphPlan(air_cargo()).execute()
for step, action in enumerate(linearize(plan), 1):
    print(f"{step}. {action}")
```
