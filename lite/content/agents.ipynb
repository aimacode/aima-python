## The two-square vacuum world

A simple reflex agent senses only its current square; a model-based agent remembers which squares it has already cleaned. We compare their average score over random starting configurations.

```python
from aima.agents import (TrivialVacuumEnvironment, ReflexVacuumAgent,
                         ModelBasedVacuumAgent, compare_agents)

scores = compare_agents(TrivialVacuumEnvironment,
                        [ReflexVacuumAgent, ModelBasedVacuumAgent],
                        n=10, steps=100)
for agent, score in scores:
    print(f"{agent.__name__:24s} average score: {score:6.1f}")
```
