# aima-python in the browser

This is a [JupyterLite](https://jupyterlite.readthedocs.io) site running
entirely in your browser via [Pyodide](https://pyodide.org) — there is no
server. It is a proof of concept for a Python web companion to *Artificial
Intelligence: A Modern Approach* (see
[issue #1072](https://github.com/aimacode/aima-python/issues/1072)).

Every notebook starts by installing the `aima` package into the browser kernel
with the cell below. Pyodide already provides numpy/scipy/matplotlib/networkx,
so `aima`'s heavy native dependencies (TensorFlow/Keras, OpenCV, cvxopt) are
skipped with `deps=False`; the demo notebooks use only the lightweight modules.

Open **search.ipynb** and **games.ipynb** from the file browser on the left.
