# Search in the browser

Uninformed and informed search on the Romania map from chapter 3, running
fully in your browser with `aima.search`. Run the cells top to bottom.

```python
import piplite
await piplite.install("aima", deps=False)
```

```python
from aima.search import GraphProblem, romania_map
from aima.search import breadth_first_graph_search, uniform_cost_search, astar_search

problem = GraphProblem("Arad", "Bucharest", romania_map)

for name, algorithm in [
    ("Breadth-first", breadth_first_graph_search),
    ("Uniform-cost", uniform_cost_search),
    ("A* (straight-line heuristic)", astar_search),
]:
    goal = algorithm(problem)
    print(f"{name:30s} {['Arad'] + goal.solution()}  (cost {goal.path_cost})")
```

A* finds the optimal route Arad → Sibiu → Rimnicu → Pitesti → Bucharest
(cost 418), while breadth-first returns a shorter-in-hops but costlier path.
