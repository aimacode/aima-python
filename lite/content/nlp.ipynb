## CYK parsing

Given a probabilistic grammar in Chomsky normal form, the CYK algorithm fills a chart of the most likely subtrees spanning each substring.

```python
from aima import nlp
from aima.nlp import CYK_parse

words = ['the', 'robot', 'is', 'good']
chart = CYK_parse(words, nlp.E_Prob_Chomsky)
print('parsing:', ' '.join(words))
print('chart holds', len(chart), 'scored (symbol, span) entries')

# most likely category spanning the whole sentence
n = len(words)
spanning = {sym: p for (sym, start, length), p in chart.items() if start == 0 and length == n}
best = max(spanning, key=spanning.get)
print('best category for the full span:', best, 'with probability', round(spanning[best], 6))
```
