## Counting n-grams

We build word models from a short inline corpus and inspect the most frequent words and word pairs.

```python
from aima.text import UnigramWordModel, NgramWordModel, words

corpus = words(
    'the quick brown fox jumps over the lazy dog '
    'the fox is quick and the dog is lazy '
    'the quick dog jumps over the lazy fox')

unigram = UnigramWordModel(corpus)
bigram = NgramWordModel(2, corpus)
print('most frequent words: ', unigram.top(3))
print('most frequent bigrams:', bigram.top(3))
print("P('quick') =", round(unigram['quick'], 4))
```
