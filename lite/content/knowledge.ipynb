## Learning a concept from examples

Should this animal stay dry? Each example is described by attributes and a boolean `GOAL`; `current_best_learning` searches for a hypothesis (a disjunction of attribute constraints) consistent with all of them.

```python
from aima.knowledge import current_best_learning, guess_value

examples = [
    {'Species': 'Cat', 'Rain': 'Yes', 'Coat': 'No',  'GOAL': True},
    {'Species': 'Cat', 'Rain': 'Yes', 'Coat': 'Yes', 'GOAL': True},
    {'Species': 'Dog', 'Rain': 'Yes', 'Coat': 'Yes', 'GOAL': True},
    {'Species': 'Dog', 'Rain': 'Yes', 'Coat': 'No',  'GOAL': False},
    {'Species': 'Dog', 'Rain': 'No',  'Coat': 'No',  'GOAL': False},
    {'Species': 'Cat', 'Rain': 'No',  'Coat': 'No',  'GOAL': False},
    {'Species': 'Cat', 'Rain': 'No',  'Coat': 'Yes', 'GOAL': True}]

h = current_best_learning(examples, [{'Species': 'Cat'}])
predictions = [guess_value(e, h) for e in examples]
actual = [e['GOAL'] for e in examples]
print('predictions:', predictions)
print('all correct:', predictions == actual)
```
