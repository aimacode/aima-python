# Learning from examples in the browser

Decision-tree and naive-Bayes classifiers (chapters 19-20) running fully in your browser with `aima.learning`. We build a small dataset inline so no data files are needed.

```python
import piplite
await piplite.install("aima", deps=False)
```

## The "play tennis" dataset

A classic toy dataset: predict whether to play tennis from the weather.

```python
from aima.learning import DataSet, DecisionTreeLearner
from aima.probabilistic_learning import NaiveBayesLearner

examples = [
    ['Sunny', 'Hot', 'High', 'Weak', 'No'], ['Sunny', 'Hot', 'High', 'Strong', 'No'],
    ['Overcast', 'Hot', 'High', 'Weak', 'Yes'], ['Rain', 'Mild', 'High', 'Weak', 'Yes'],
    ['Rain', 'Cool', 'Normal', 'Weak', 'Yes'], ['Rain', 'Cool', 'Normal', 'Strong', 'No'],
    ['Overcast', 'Cool', 'Normal', 'Strong', 'Yes'], ['Sunny', 'Mild', 'High', 'Weak', 'No'],
    ['Sunny', 'Cool', 'Normal', 'Weak', 'Yes'], ['Rain', 'Mild', 'Normal', 'Weak', 'Yes'],
    ['Sunny', 'Mild', 'Normal', 'Strong', 'Yes'], ['Overcast', 'Mild', 'High', 'Strong', 'Yes'],
    ['Overcast', 'Hot', 'Normal', 'Weak', 'Yes'], ['Rain', 'Mild', 'High', 'Strong', 'No']]
tennis = DataSet(name='tennis', examples=examples,
                 attr_names='Outlook Temp Humidity Wind PlayTennis', target='PlayTennis')
```

## Decision tree vs. naive Bayes

```python
tree = DecisionTreeLearner(tennis)
bayes = NaiveBayesLearner(tennis, continuous=False)

for day in [['Overcast', 'Hot', 'Normal', 'Weak'],
            ['Sunny', 'Hot', 'High', 'Strong'],
            ['Rain', 'Cool', 'Normal', 'Weak']]:
    print(day, '-> tree:', tree(day), '| naive Bayes:', bayes(day))
```
