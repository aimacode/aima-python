## Propositional logic

Model checking by truth-table enumeration, and satisfiability via DPLL.

```python
from aima.logic import expr, tt_entails, dpll_satisfiable

# modus ponens: does  P & (P => Q)  entail  Q ?
print("(P & (P ==> Q)) |= Q :", tt_entails(expr("(P & (P ==> Q))"), expr("Q")))

# satisfiability
print("P | ~P satisfiable   :", bool(dpll_satisfiable(expr("P | ~P"))))
print("P & ~P satisfiable   :", bool(dpll_satisfiable(expr("P & ~P"))))
```

## First-order logic

The classic "West is a criminal" knowledge base (section 9.3), solved by
forward chaining.

```python
from aima.logic import expr, FolKB, fol_fc_ask

clauses = [
    "(American(x) & Weapon(y) & Sells(x, y, z) & Hostile(z)) ==> Criminal(x)",
    "Owns(Nono, M1)",
    "Missile(M1)",
    "(Missile(x) & Owns(Nono, x)) ==> Sells(West, x, Nono)",
    "Missile(x) ==> Weapon(x)",
    "Enemy(x, America) ==> Hostile(x)",
    "American(West)",
    "Enemy(Nono, America)",
]
kb = FolKB([expr(c) for c in clauses])
print("Who is a criminal? ", list(fol_fc_ask(kb, expr("Criminal(x)"))))
```
